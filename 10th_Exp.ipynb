```python
import numpy as np
from sklearn.mixture import GaussianMixture

# Sample dataset
X = np.array([
    [1, 2],
    [1.5, 1.8],
    [5, 8],
    [8, 8],
    [1, 0.6],
    [9, 11]
])

# Gaussian Mixture Model
gmm = GaussianMixture(n_components=2, random_state=42)

# Train model
gmm.fit(X)

# Predict clusters
labels = gmm.predict(X)

print("Data Points:")
print(X)

print("\nCluster Labels:")
print(labels)
```
